```python
# img = cv2.imread('/content/gdrive/My Drive/Colabip/Images/fce(salt_pepper noise).bmp')
img = cv2.imread('/content/gdrive/My Drive/Colabip/Images/home3.jpg')
gimg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

cv2_imshow(gimg)
```

```python
z = gimg.reshape((-1,3))
print(z.shape)
```

```python
z = np.float32(z)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
k = 8
ret, label, center = cv2.kmeans(z, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
res2 = res.reshape((gimg.shape))
cv2_imshow(res2)
```
